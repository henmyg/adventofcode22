```python
# constants
isTest = False
isTwo = True
step = 4
if isTest:
    filename = "test"
    stackCount = 3
    lastStackLine = 3
    firstMoveLine = 5
else:
    filename = "input"
    stackCount = 9
    
    lastStackLine = 8
    firstMoveLine = 10
```

```python
with open(filename) as file:
    lines = file.readlines()

    stacks = []

    line = lines[0]
    for i in range(0,stackCount):
        stacks.append([])
    
    line = lines[0] # TODO

    for line in lines[:lastStackLine]:
        for i in range(0, stackCount):
            index = i*step
            if line[index] == '[':
                stacks[i].append(line[index+1])

    stacks = list( map(lambda s: list(reversed(s)), stacks))
```

```python
moveLines = lines[firstMoveLine:]

for line in moveLines:
    line = list(reversed(line.replace("move ", "").replace(" from ", "").replace(" to ", "")))
    to = int(line[1])-1
    frm = int(line[2])-1
    count = int("".join(list(reversed(line[3:]))))

    if isTwo:
        print("")
        crates = stacks[frm][-count:]
        stacks[frm] = stacks[frm][:-count]
        stacks[to].extend(crates)

    else:
        for i in range(0, count):
            crate = stacks[frm].pop()
            stacks[to].append(crate)
```

```python
resultOne = "".join(list(map(lambda s: s[-1], stacks)))
print(resultOne)
```
